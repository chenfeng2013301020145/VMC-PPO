```python
import pandas as pd
import numpy as np
import os

import matplotlib.pyplot as plt

def plot_training(L, W, lattice_type='sq',version='1', fn_type=None, show=True, true_value=-1):
    if fn_type == None:
        fn = 'results/HS_2d_%s_L%dW%d_vmcppo%s'%(lattice_type,L,W,version)
    else:
        fn = 'results/HS_2d_%s_L%dW%d_vmcppo%s_%s'%(lattice_type,L,W,version,fn_type)
    txt_fn = os.path.join(fn, 'exp_log.txt')

    df = pd.read_table(txt_fn)
    data = np.array(df)

    epoch = []
    E = []
    mE = []
    cmE = []
    stdE = []
    dfs = []
    sps = []
    sym_sps = []

    start_label = len(data) + 1000
    for idx, d in enumerate(data):
        strings = d[0].split(" ")
        if len(strings) > 2 and strings[-2] == 'Start':
            start_label = idx
        if len(strings) > 2 and strings[-2] == 'Finish':
            break

        if idx > start_label:
            epoch.append(float(strings[3].split(",")[0]))
            E.append(float(strings[5].split(",")[0]))
            mE.append(float(strings[7].split(",")[0]))
            cmE.append(float(strings[9].split(",")[0]))
            stdE.append(float(strings[11].split(",")[0]))
            dfs.append(float(strings[13].split(",")[0]))
            sps.append(float(strings[19].split(",")[0]))
            #if version == '2':
            sym_sps.append(float(strings[21].split(",")[0]))
        
    E = np.array(E)
    stdE = np.array(stdE)

    with plt.style.context('seaborn'):
        fig, axs = plt.subplots(2, 2, sharey=False, figsize=(16,6))
        ax1,ax2,ax3,ax4 = axs[0][0],axs[1][0],axs[0][1],axs[1][1]


        ax1.plot(epoch, E, label='E')
        ax1.plot(epoch, mE, ':', label='mE')
        ax1.plot(epoch, cmE, ':', label='cmE')
        #ax1.fill_between(epoch, E-stdE, E+stdE, alpha=0.2, zorder=-1)
        ax1.set_ylabel(r'$E/N$')
        ax1.legend()

        ax2.semilogy(epoch, abs(stdE/E), label=r'$\Delta E/E$')
        ax2.semilogy(epoch, abs(E - true_value)/abs(true_value), label=r'$|E - E_{ED}|/|E_{ED}|$')
        ax2.set_ylabel(r'$\Delta E/E$')
        ax2.set_xlabel('epoch')
        ax2.legend()

        ax3.semilogy(epoch, dfs)
        ax3.set_ylabel('FS distance')
        #ax3.set_xlabel('epoch')

        #ax4.plot(E, sps, '.')
        #if version == '2':
        ax4.plot(epoch, sym_sps, label='sym')
        #else:
        ax4.plot(epoch, sps, label='int')
        ax4.set_ylabel('number of samples')
        ax4.set_xlabel('epoch')
        ax4.legend()

        # if version == '2':
        #     ax5 = ax4.twinx()
        #     ax5.plot(epoch, sym_sps, color='tab:green', label='sym')
        #     ax5.tick_params(axis='y', color='tab:green')
            

        if show:
            plt.show()
```
